```python
import pandas as pd
from sklearn import preprocessing


# Encode the label by threshold 1400
label_encoder = preprocessing.LabelEncoder()
popular_label = pd.Series(label_encoder.fit_transform(popularity >= 1400))
```

```python
# Get the features from dataset
features_raw = data.drop(['url',data.keys()[1],data.keys()[-1]], axis=1)
display(features_raw.head())
```

```python
# Normalize the numerical features
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

numerical = [' n_tokens_title', ' n_tokens_content', ' num_hrefs', ' num_self_hrefs', ' num_imgs',' num_videos',\
            ' average_token_length',' num_keywords',' self_reference_min_shares',' self_reference_max_shares',\
             ' self_reference_avg_sharess']
features_raw[numerical] = scaler.fit_transform(data[numerical])
display(features_raw.head(n = 1))
```

```python
import matplotlib.pyplot as plt
import numpy as np

columns_day = features_raw.columns.values[29:36]
unpop = data[data[' shares'] < 1400]
pop = data[data[' shares'] >= 1400]
unpop_day = unpop[columns_day].sum().values
pop_day = pop[columns_day].sum().values

plt.figure(figsize=(13, 5))
plt.title("Count of popular/unpopular news over different days of the week", fontsize=16)
plt.bar(np.arange(len(columns_day)), pop_day, width=0.3, align="center", color='cyan', label="popular")
plt.bar(np.arange(len(columns_day)) - 0.3, unpop_day, width=0.3, align="center", color='goldenrod', label="unpopular")
plt.xticks(np.arange(len(columns_day)), columns_day)
plt.ylabel("Count", fontsize=12)
plt.xlabel("Days of the week", fontsize=12)
plt.legend(loc="upper right")
plt.tight_layout()
plt.savefig("days.pdf")
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Perform feature selection with SelectKBest
selector = SelectKBest(score_func=f_classif, k=30)
selected_features = selector.fit_transform(features_raw, popular_label)

# Get the selected feature indices
selected_feature_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_feature_names = features_raw.columns[selected_feature_indices]

# Print the selected feature names
print("Selected Features:")
print(selected_feature_names)
```

```python
# Split data into training and testing sets

from sklearn.model_selection import train_test_split

# Split the data
x_train, x_test, y_train, y_test = train_test_split(selected_features, popular_label , test_size=0.1, random_state=0)

# Print the number of training and test samples
print("Number of training samples:", x_train.shape[0])
print("Number of test samples:", x_test.shape[0])
print()
```

```python
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
rfc = RandomForestClassifier(n_estimators = 50, random_state = 2)
lr = LogisticRegression(solver = 'liblinear')
knn = KNeighborsClassifier()
dtc = DecisionTreeClassifier(max_depth = 5)
adc = AdaBoostClassifier(n_estimators = 50, random_state = 2)
bc = BaggingClassifier(n_estimators = 50, random_state = 2)
```

```python
classifier = {
    'RandomForestClassifier' : rfc,
    'Logistic_Regression': lr,
    'KNN': knn,
    'Decision Tree': dtc,
    'AdaBoost': adc,
    'Bagging': bc,
}
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

def train_classifier(classifier, x_train, x_test, y_train, y_test):
    classifier.fit(x_train, y_train)
    y_pred = classifier.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred)
    return accuracy, f1, auc
```

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score

accuracy_score1 = []
f1_score3 = []
auc_score4 = []

for name, clsr in classifier.items():
    current_accuracy, current_f1_score, current_auc_score = train_classifier(clsr, x_train, x_test, y_train, y_test)
    print("Model -", name)
    print("Accuracy -", current_accuracy)
    print("F1-score -", current_f1_score)
    print("AUC -", current_auc_score)
    print("\n")

    accuracy_score1.append(current_accuracy)
    f1_score3.append(current_f1_score)
    auc_score4.append(current_auc_score)
```

```python
performance_dp = pd.DataFrame({'Algorithm': classifier.keys(),
                               'Accuracy': accuracy_score1,
                               'F1-score': f1_score3,
                               'AUC': auc_score4})
performance_dp
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the performance metrics and classifiers
metrics = ['Accuracy', 'F1-score', 'AUC']
classifiers = performance_dp['Algorithm']

# Create a figure with subplots
fig, axes = plt.subplots(nrows=len(metrics), ncols=1, figsize=(15, 15))

# Define colors for each classifier
colors = ['#A00000', '#00A0A0', '#00A000', '#0000A0', '#A000A0', '#FF8000']

# Iterate over the metrics
for i, metric in enumerate(metrics):
    # Plot the bar graph for the metric
    axes[i].bar(classifiers, performance_dp[metric], color=colors)
    axes[i].set_ylabel(metric)
    axes[i].set_ylim([0, 1])  # Set the y-axis limits for better comparison

    # Add data labels to the bars
    for j, v in enumerate(performance_dp[metric]):
        axes[i].text(j, v + 0.01, f'{v:.2f}', ha='center')

    # Set the color of the y-axis labels
    axes[i].yaxis.label.set_color('black')

    # Set the color of the tick labels
    axes[i].tick_params(colors='black')

# Set the x-axis label and title
axes[-1].set_xlabel('Algorithm')
plt.suptitle('Performance Metrics of Different Classifiers')

# Create custom legend handles
legend_handles = [plt.Rectangle((0, 0), 1, 1, color=color) for color in colors]
legend_labels = classifiers

# Add legend
plt.legend(legend_handles, legend_labels, loc='upper right')

# Adjust the spacing between subplots
plt.subplots_adjust(hspace=0.5)

# Show the plot
plt.show()
```
